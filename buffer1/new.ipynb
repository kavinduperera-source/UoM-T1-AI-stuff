```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
from scipy.signal import savgol_filter
```

```python
file_path = r"C:\Users\acer\Desktop\signal.csv"
df = pd.read_csv(file_path, header=None, names =['Signal'])
x = range(len(df))
```

```python
signal_array = df['Signal'].to_numpy()
window_size = 21
'''moving average smoothing'''
smoothed_signal = np.convolve(signal_array, np.ones(window_size)/window_size, mode='same')
```

```python
plt.plot(x,smoothed_signal)
plt.grid(True)
plt.show()
```

```python
signal_gaussian = gaussian_filter1d(signal_array, sigma=3)
plt.plot(x, signal_gaussian)
plt.grid(True)
plt.show()
```

```python
signal_sg = savgol_filter(signal_array, window_length=21, polyorder=3)
plt.plot(x, signal_sg) 
plt.grid(True)
plt.show()
```

```python
'''Lets visualize the given signal data'''
x = range(len(df))
y= df['Signal']
plt.figure(figsize=(10,5))
plt.plot(x,y, label='Signal')
plt.xlabel('Domain(time)')
plt.ylabel('function value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sin_amplitude = 1.0
sq_amplitude = 0.5
sin_frequency = 0.02
sq_frequency = 0.01
phase = 0.5
#sampling_rate = 100
duration = 4

x = np.arange(0,4000) 

sine_wave = sin_amplitude * np.sin(sin_frequency * x)
square_wave = sq_amplitude * np.sign(np.sin(sq_frequency * x+phase))
plt.figure(figsize=(10, 5))
plt.plot(x, sine_wave, label='Sine Wave')
plt.plot(x, square_wave, label='Square Wave')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Sine and Square Waves')
plt.legend()
plt.grid(True)
plt.show()
```

```python
z = square_wave + sine_wave
plt.figure(figsize=(10, 5))
plt.plot(x, z, label='Combined Wave', color='green')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Combined Sine and Square Wave')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the base domain
x = np.arange(4001)  # 0 to 4000

# Create sine + square mixture function
sine_component = np.sin(x * 0.02)  # Sine wave
square_component = np.sign(np.sin(x * 0.01))  # Square wave
y_clean = sine_component + 0.6 * square_component  # Mixed signal

# Add Gaussian noise to all points
noise = np.random.normal(0, 0.15, size=len(x))  # Mean=0, Std=0.2
y_noisy = y_clean + noise

# Plot comparison
plt.figure(figsize=(14, 8))

plt.subplot(2, 1, 1)
plt.plot(x, y_clean, 'b-', linewidth=1.5, alpha=0.8, label='Clean Signal')
plt.title('Clean Sine + Square Mixture')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 1, 2)
plt.plot(x, y_noisy, 'r-', linewidth=1, alpha=0.7, label='Noisy Signal')
plt.title('With Gaussian Noise Added')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
